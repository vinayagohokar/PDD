```python
from ultralytics import YOLO
# Load the pre-trained YOLOv11 Nano model
model = YOLO("yolo11n.pt")  # YOLOv11 Nano weights

# Train the model
model.train(
    data="/content/data/data.yaml",  # Path to dataset configuration file
    epochs=50,               # Number of epochs
    batch=16,                # Batch size
    imgsz=640,               # Image size
    name="train",  # Experiment name
    device=0                 # Use GPU (0) or CPU (-1)
)
```

```python
metrics = model.val(
    data="/content/data/data.yaml",  # Path to dataset configuration file
    imgsz=640                # Image size
)
print(metrics)
```

```python
model.export(format="tflite")
```
